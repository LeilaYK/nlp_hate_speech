##### 데이터 분할

````python
X = df['comments']
y = df['label']
````

#### 1. 백터화
- TFIDF
- CountVectorizer 줄여서 CV

#### 1.1 cv

````python
from sklearn.feature_extraction.text import CountVectorizer

vec_cv = CountVectorizer(analyzer = "word", max_features = 5000) 

X_cv = vec_cv.fit_transform(X)
````

````python
X_cv.shape
````

#### 1.2 tfidf
- min_df : 
```
이 설정 값보다 작은 df(해당 단어가 발생하는 document 수)라면 벡터화 과정에서 제거한다는 의미로, 전체 문서에서 최소한 이정도는 발생해야 의미있는 정보를 얻을수 있고, 한번 정도밖에 나오지 않은 단어는 판단하기보단 제외시켜놓는다는 의미입니다.
```
- analyzer : word와 char로 2가지 옵션을 제공하는데, word는 단어 하나를 단위로, char은 문자 하나의 단위로 TF-IDF를 구한다는 뜻입니다.
- sublinear_tf : TF(문서의 단어 발생 빈도)의 스무딩 여부
- ngram_range : 빈도의 기본 단위를 어느 범위의 n-gram으로 설정할 것인지를 보는 인자
- max_features : 각 벡터의 최대 길이, 특징의 길이입니다.

````python
from sklearn.feature_extraction.text import TfidfVectorizer

vec_tf = TfidfVectorizer(min_df = 0.0, analyzer = 'char', sublinear_tf=True, 
    ngram_range = (1,3), max_features=5000)


X_tf = vec_tf.fit_transform(X)
````

````python
X_tf.shape
````

````python
vec_tf.get_feature_names()
````

#### 2. Logistic Regression 모델 적용
- 모델 검증
- 데이터가 독립적이고 동일한 분포를 가진 경우
    - KFold, RepeatedKFold, LeaveOneOut(LOO), LeavePOutLeaveOneOut(LPO)
- 동일한 분포가 아닌 경우
    - StratifiedKFold, RepeatedStratifiedKFold, StratifiedShuffleSplit

````python
from sklearn.model_selection import KFold
from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import LogisticRegression

# KFold & StratifiedKFold의 객체 선언
kf = KFold(n_splits=10)
skfold = StratifiedKFold(n_splits=10)

# LogisticRegression의 객체 선언
# lgs = LogisticRegression(class_weight = 'balanced', random_state=13)
lgs = LogisticRegression(multi_class='multinomial', class_weight = 'balanced', random_state=13)
````

````python
# kfold 검증 코드 함수화
def get_acc_kf(data):
    lg_accuracy = []
    for train_idx, test_idx in kf.split(data):
        X_train, X_test = data[train_idx], data[test_idx]
        y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]
        
        lgs.fit(X_train, y_train)
        pred = lgs.predict(X_test)
        
        lg_accuracy.append(accuracy_score(y_test, pred))
        print(np.unique(pred, return_counts=True))
    return lg_accuracy
````

##### KFold 검증

````python
from sklearn.metrics import (accuracy_score, recall_score, precision_score, 
                             roc_auc_score, f1_score)
# tfidf vectorizer
lg_tf_acc = get_acc_kf(X_tf)
# count vectorizer
lg_cv_acc = get_acc_kf(X_cv)
````

##### StratifiedKFold 검증

````python
from sklearn.model_selection import cross_val_score

lg_tf_acc = cross_val_score(lgs, X_tf, y, scoring=None, cv=skfold)
lg_cv_acc = cross_val_score(lgs, X_cv, y, scoring=None, cv=skfold)
````

````python
print("StratifiedKFold :: TF Acc : ", np.mean(lg_tf_acc), ", CV Acc : ", np.mean(lg_cv_acc)) 
````
